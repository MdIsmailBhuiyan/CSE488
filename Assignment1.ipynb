# Finding Frequent ItemSets Using MRJob

```python
%%writefile FrequentItemsets1.py
from mrjob.job import MRJob
from mrjob.step import MRStep
import csv

class FrequentItemsets1(MRJob):
    # Threshold for support
    support_threshold = 5

    # Define support-Treshold while running
    def configure_args(self):
        super(FrequentItemsets1, self).configure_args()
        self.add_passthru_arg('--support-threshold', type=int, default=5, help='Minimum support threshold')

    def steps(self):
        return [
            MRStep(mapper=self.mapper1,
                   reducer=self.reducer1),
            MRStep(mapper=self.mapper2,
                   reducer=self.reducer2)
        ]
    def mapper1(self, _, line):

        if line.startswith("Invoice"):  # Skip the header row
            return

        fields = line.strip().split(',')

        invoice = fields[0]
        description = fields[2].strip()  # strip to remove extra spaces

        # Group items by Invoice ID.
        yield invoice, description

    def reducer1(self, invoice, items): #reducer_group_items
        # Make a list of items per invoice
        basket = list(items)
        yield None, basket

    def mapper2(self, _, basket): #mapper_count_itemsets
        # Generate and count all pairs of items in a basket.
        for i in range(len(basket)):
            for j in range(i + 1, len(basket)):
                itemset = tuple(sorted([basket[i], basket[j]]))
                yield itemset, 1

    def reducer2(self, itemset, counts):
        #Reducer to calculate the support of itemsets and filter by the threshold.
        support = sum(counts)
        if support >= self.options.support_threshold:
            yield itemset, support


if __name__ == '__main__':
    FrequentItemsets1.run()
```

# Finding the total quantity of the items.

```python
%%writefile SumQuantities.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class SumQuantities(MRJob):

    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer)
        ]

    def mapper(self, _, line):
        if line.startswith("Invoice"):
            return  # Skip the header line

        fields = line.strip().split(',')

        # Ensure there are enough columns
        if len(fields) < 4:
            return

        description = fields[2].strip()  # Strip spaces from description
        quantity = int(fields[3].strip())  # Convert Quantity to integer
        yield description, quantity

    def reducer(self, description, quantities):
        total_quantity = sum(quantities)
        yield description, total_quantity

if __name__ == '__main__':
    SumQuantities.run()
```

```python
single_item_count = {}
pair_item_count = {}

with open('results1.txt', 'r') as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) == 2:
            item, count = parts
            item = item.strip('"')

            if ',' in item:
                items = item.split(', ')
                if len(items) == 2:
                    item1, item2 = items
                    pair_item_count[(item1, item2)] = int(count)
            else:
                single_item_count[item] = int(count)

item_name1 = input("Enter 1st item name: ")
item_name2 = input("Enter 2nd item name: ")

confidence_pair = pair_item_count[(item_name1, item_name2)] / single_item_count[item_name1]
print(f'Confidence of {item_name1} => {item_name2}: {confidence_pair:.2f}')
```
